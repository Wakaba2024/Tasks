## DATA CLEANING AND TRANSFORMATION

```python
retail_df.info() #Hepls know the data types of the different columns
```

```python
# Categorical Exploration: Write a loop that prints the unique values for each object-type column (excluding Order Date).
for col in retail_df.select_dtypes(include=['object']).columns: # It loops through object-type columns
        if col != 'Order Date': # Skips the specified column
            uniques = retail_df[col].unique()  # Gets unique values in the column
            print(f"Unique values for column '{col}' ({len(uniques)}):") # Outputs the column name and the number of unique values
            print(uniques) # Outputs a list of all unique values
            print("-" * 20) # Enhances understandability by printing a separator line
```

```python
# Replace Invalids: In pandas, replace the entries 'Not Available' and 'unknown' in the Ship Mode column with NaN.
retail_df['Ship Mode'] = retail_df['Ship Mode'].replace(['Not Available', 'unknown'], np.nan)
```

```python
retail_df.head(30) # Have a preview of the changed entries
```

```python
# Rename Columns: Show how to rename all DataFrame columns by lowercasing and replacing spaces with underscores.
retail_df.columns = retail_df.columns.str.lower().str.replace(' ', '_')
```

```python
retail_df.head(20)
```

```python
# Create Derived Columns:
# Compute a discount column as list_price * discount_percent / 100.
retail_df['discount'] = retail_df['list_price'] * retail_df['discount_percent']/100
# Compute a selling_price column as list_price - discount.
retail_df['selling_price'] = retail_df['list_price'] - retail_df['discount']
# Compute a profit column as selling_price - cost_price.
retail_df['profit'] = retail_df['selling_price'] - retail_df['cost_price']
```

```python
retail_df.head()
```

```python
# Drop Unneeded Columns: Remove cost_price, list_price, and discount_percent from the DataFrame.
retail_df.drop(['cost_price', 'list_price', 'discount_percent'], axis=1, inplace=True)
```

```python
retail_df.head()
```

```python
# Datetime Conversion: Convert the order_date column to pandas datetime with ISO8601 format
retail_df['order_date'] = pd.to_datetime(retail_df['order_date'], format='%Y-%m-%d')
```

```python
# Convert the dataframe to a csv
retail_df.to_csv('retails.csv', index = False)
```

```python
conn = psycopg2.connect(database = 'retail_db',
                        user = 'pytest',
                        host = 'localhost',
                        password = '1234',
                        port = 5432)
```

```python
# Identify who is the current user after creating the connection
cur = conn.cursor()
cur.execute("SELECT current_user;")
print("Current PostgreSQL user:", cur.fetchone()[0])
```

```python
# Granting permissions to the user and creating the schema
cur = conn.cursor()

cur.execute("CREATE SCHEMA IF NOT EXISTS pysql_test;")
cur.execute("GRANT ALL ON SCHEMA pysql_test TO pytest;")
cur.execute("ALTER ROLE pytest SET search_path = pysql_test, public;")

conn.commit()
```

```python
# Create table orders
cur = conn.cursor()

cur.execute("""
CREATE TABLE pysql_test.orders (
    order_id INT PRIMARY KEY,
    order_date DATE NOT NULL,
    ship_mode VARCHAR(100),
    segment VARCHAR(100) NOT NULL,
    country VARCHAR(100) NOT NULL,
    city VARCHAR(100) NOT NULL,
    state VARCHAR(100) NOT NULL,
    postal_code INT NOT NULL,
    region VARCHAR(100) NOT NULL,
    category VARCHAR(100) NOT NULL,
    sub_category VARCHAR(100) NOT NULL,
    product_id VARCHAR(100) NOT NULL,
    quantity INT NOT NULL,
    discount NUMERIC(10, 2) NOT NULL,
    selling_price NUMERIC(10, 2) NOT NULL,
    profit NUMERIC(10, 2) NOT NULL
);
""")

conn.commit()
```

```python
#  Load and copy the CSV (without the header) into the PostgreSQL table
cur = conn.cursor()

with open(r"C:\Users\user\PYTHON_JOHN\retails.xls", 'r') as f:
    next(f)  # skip the header
    cur.copy_expert("COPY orders FROM STDIN WITH CSV", f)

conn.commit()
cur.close()
# r'...' treats the file path as a raw string, avoiding issues with backslashes (\).

# next(f) skips the header row of the CSV file.

# cur.copy_expert(...) runs the PostgreSQL COPY command,

# Efficiently loads the data in bulk from the file into the orders table.

# STDIN reads the file directly from the Python program.

# WITH CSV tells PostgreSQL to interpret the input as CSV-formatted data.
```

```python
cur = conn.cursor()
cur.execute('SELECT * FROM orders;')
rows = cur.fetchall()
conn.commit()
conn.close()
for row in rows:
    print(row)
```

## PYTHON CODE

```python
# Total Revenue per Category: Sum selling_price grouped by category, ordered descending.
print(retail_df.groupby("category")["selling_price"].sum().sort_values(ascending=False))
```

```python
#Top 3 Profitable Cities: Identify the three cities with the highest total profit.
print(retail_df.groupby("city")["profit"].sum().sort_values(ascending=False).head(3))
```

```python
# Profit Margin per Product: Compute (SUM(profit)/SUM(selling_price))*100 per product_id, list the top 25.
retail_df_margin = retail_df.groupby("product_id").agg({"profit": "sum", "selling_price": "sum"})
retail_df_margin["margin_percent"] = (retail_df_margin["profit"] / retail_df_margin["selling_price"]) * 100
print(retail_df_margin.sort_values("margin_percent", ascending=False).head(25))
```

```python
# Order Counts by Shipping Mode: Count order_id per ship_mode, ordered descending
print(retail_df["ship_mode"].value_counts())
```

```python
# Month with Highest Orders: Find the month number (MONTH(order_date)) with the greatest count of order_id
print(retail_df["order_date"].dt.month.value_counts().idxmax())
```

```python
# Top 5 Most Discounted Products: Select product_id and discount, ordered by discount descending.
print(retail_df[["product_id", "discount"]].sort_values(by="discount", ascending=False).head(5))
```

```python
# Top 3 Cities by Sales: Sum selling_price per city, list the top three
print(retail_df.groupby("city")["selling_price"].sum().sort_values(ascending=False).head(3))
```

```python
# Average Profit for 2023: Calculate the average profit for orders where YEAR(order_date)=2023
print(retail_df[retail_df["order_date"].dt.year == 2023]["profit"].mean())
```

```python
# Sub-Category Sales Ranking: Sum selling_price by sub_category, show the top five.
print(retail_df.groupby("sub_category")["selling_price"].sum().sort_values(ascending=False).head(5))
```

```python
# Average Selling Price per Category: Compute average selling_price for each category, ordered descending.
print(retail_df.groupby("category")["selling_price"].mean().sort_values(ascending=False))
```

```python
# . Top 10 Products by Order Count: Count order_id per product_id, list the top ten.
print(retail_df["product_id"].value_counts().head(10))
```

```python
# Top 5 Profitable Products: Sum profit per product_id, list the top five.
print(retail_df.groupby("product_id")["profit"].sum().sort_values(ascending=False).head(5))
```

```python
# Year with Highest Sales: Find YEAR(order_date) with maximum sum of selling_price.
print(retail_df.groupby(retail_df["order_date"].dt.year)["selling_price"].sum().idxmax())
```

```python
# Region with Lowest Discount: Identify region with the lowest average discount
print(retail_df.groupby("region")["discount"].mean().sort_values().head(1))
```

```python
# Yearly Sales Growth: Show total sales (SUM(selling_price)) per year, ordered by year.
print(retail_df.groupby(retail_df["order_date"].dt.year)["selling_price"].sum())
```

```python
# Profit Contribution by Category: For each category, compute SUM(profit)*100 / (SELECT SUM(profit) FROM orders).
total_profit = retail_df["profit"].sum()
print((retail_df.groupby("category")["profit"].sum() * 100 / total_profit))
```

```python
# Most Common Shipping Mode: Find the single ship_mode used by the most orders.
print(retail_df["ship_mode"].value_counts().idxmax())
```

```python
#  Highest Average Order Value by Region: Determine which region has the highest average selling_price.
print(retail_df.groupby("region")["selling_price"].mean().sort_values(ascending=False).head(1))
```

```python
# Category & Sub-Category Sales: Sum selling_price for each (category, sub_category) pair.
print(retail_df.groupby(["category", "sub_category"])["selling_price"].sum())
```

```python
# Top 3 Profitable Products per Sub-Category: For each sub_category, list the top three product_id by total profit.
top3_per_sub = retail_df.groupby(["sub_category", "product_id"])["profit"].sum().reset_index()
top3 = top3_per_sub.groupby("sub_category").apply(lambda x: x.nlargest(3, "profit")).reset_index(drop=True)
print(top3)
```

```python
# Monthly Order Counts for 2023: Count order_id by MONTH(order_date) for YEAR(order_date)=2023, list in descending order.
print(retail_df[retail_df["order_date"].dt.year == 2023].groupby(retail_df["order_date"].dt.month)["order_id"].count().sort_values(ascending=False))
```

```python
# States with Lowest Sales: Identify the ten state values with the smallest total selling_price
print(retail_df.groupby("state")["selling_price"].sum().sort_values().head(10))
```

```python
# Top 5 States by Revenue: Find the five state values contributing most to total revenue.
print(retail_df.groupby("state")["selling_price"].sum().sort_values(ascending=False).head(5))
```

```python
# Average Discount by Sub-Category: Compute average discount per sub_category, ordered descending.
print(retail_df.groupby("sub_category")["discount"].mean().sort_values(ascending=False))
```

```python
# Top 10 Products by Revenue: Sum selling_price per product_id, list the top ten by total revenue.
print(retail_df.groupby("product_id")["selling_price"].sum().sort_values(ascending=False).head(10))
```

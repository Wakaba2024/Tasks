## DATA CLEANING AND TRANSFORMATION

```python
df.info() #Hepls know the data types of the different columns
```

```python
# Categorical Exploration: Write a loop that prints the unique values for each object-type column (excluding Order Date).
for col in df.select_dtypes(include=['object']).columns: # It loops through object-type columns
        if col != 'Order Date': # Skips the specified column
            uniques = df[col].unique()  # Gets unique values in the column
            print(f"Unique values for column '{col}' ({len(uniques)}):") # Outputs the column name and the number of unique values
            print(uniques) # Outputs a list of all unique values
            print("-" * 20) # Enhances understandability by printing a separator line
```

```python
# Replace Invalids: In pandas, replace the entries 'Not Available' and 'unknown' in the Ship Mode column with NaN.
df['Ship Mode'] = df['Ship Mode'].replace(['Not Available', 'unknown'], np.nan)
```

```python
df.head(30) # Have a previe of the changed entries
```

```python
# Rename Columns: Show how to rename all DataFrame columns by lowercasing and replacing spaces with underscores.
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.head(20)
```

```python
# Create Derived Columns:
# Compute a discount column as list_price * discount_percent / 100.
df['discount'] = df['list_price'] * df['discount_percent']/100
# Compute a selling_price column as list_price - discount.
df['selling_price'] = df['list_price'] - df['discount']
# Compute a profit column as selling_price - cost_price.
df['profit'] = df['selling_price'] - df['cost_price']
```

```python
df.head()
```

```python
# Drop Unneeded Columns: Remove cost_price, list_price, and discount_percent from the DataFrame.
df.drop(['cost_price', 'list_price', 'discount_percent'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Datetime Conversion: Convert the order_date column to pandas datetime with ISO8601 format
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')
```

```python
conn = psycopg2.connect(database = 'retail_db',
                        user = 'pytest',
                        host = 'localhost',
                        password = '1234',
                        port = 5432)
```

```python
# Identify who is the current user after creating the connection
cur = conn.cursor()
cur.execute("SELECT current_user;")
print("Current PostgreSQL user:", cur.fetchone()[0])
```

```python
# Granting permissions to the user and creating the schema
cur = conn.cursor()

cur.execute("CREATE SCHEMA IF NOT EXISTS pysql_test;")
cur.execute("GRANT ALL ON SCHEMA pysql_test TO pytest;")
cur.execute("ALTER ROLE pytest SET search_path = pysql_test, public;")

conn.commit()

```

```python
# Check current PostgreSQL user's privileges on all  schemas in the database.
cur = conn.cursor()
cur.execute("""
SELECT n.nspname AS schema_name,
       has_schema_privilege(n.nspname, 'CREATE') AS can_create,
       has_schema_privilege(n.nspname, 'USAGE') AS can_use
FROM pg_namespace n
WHERE n.nspname NOT LIKE 'pg_%' AND n.nspname != 'information_schema';
""")
print(cur.fetchall())


```

```python
# Create table orders
cur = conn.cursor()

cur.execute("""
CREATE TABLE pysql_test.orders (
    order_id INT PRIMARY KEY,
    order_date DATE NOT NULL,
    ship_mode VARCHAR(100),
    segment VARCHAR(100) NOT NULL,
    country VARCHAR(100) NOT NULL,
    city VARCHAR(100) NOT NULL,
    state VARCHAR(100) NOT NULL,
    postal_code INT NOT NULL,
    region VARCHAR(100) NOT NULL,
    category VARCHAR(100) NOT NULL,
    sub_category VARCHAR(100) NOT NULL,
    product_id VARCHAR(100) NOT NULL,
    quantity INT NOT NULL,
    discount NUMERIC(10, 2) NOT NULL,
    selling_price NUMERIC(10, 2) NOT NULL,
    profit NUMERIC(10, 2) NOT NULL
);
""")

conn.commit()
```

```python
#  Load and copy the CSV (without the header) into the PostgreSQL table
cur = conn.cursor()

with open(r'C:\Users\Analy\Downloads\archive\orders.csv', 'r') as f:
    next(f)  # skip the header
    cur.copy_expert("COPY orders FROM STDIN WITH CSV", f)

conn.commit()
cur.close()
# r'...' treats the file path as a raw string, avoiding issues with backslashes (\).

# next(f) skips the header row of the CSV file.

# cur.copy_expert(...) runs the PostgreSQL COPY command,

# Efficiently loads the data in bulk from the file into the orders table.

# STDIN reads the file directly from the Python program.

# WITH CSV tells PostgreSQL to interpret the input as CSV-formatted data.
```

## SQL Analysis Queries

```python
# Total Revenue per Category: Sum selling_price grouped by category, ordered descending.
cur = conn.cursor()
cur.execute('SELECT sum(selling_price) as total_rev, category FROM orders group by category order by total_rev desc;')
rows = cur.fetchall()
conn.commit()
conn.close()
for row in rows:
    print(row)
```

```python
#Top 3 Profitable Cities: Identify the three cities with the highest total profit.
cur = conn.cursor()
cur.execute('SELECT distinct(city), profit FROM orders order by profit desc limit 3;')
rows = cur.fetchall()
conn.commit()
for row in rows:
    print(row)
```

```python
# Profit Margin per Product: Compute (SUM(profit)/SUM(selling_price))*100 per product_id, list the top 25.
cur = conn.cursor()
cur.execute('SELECT product_id, ROUND((SUM(profit) / NULLIF(SUM(selling_price), 0)) * 100, 2) AS profit_margin_pct FROM orders group by product_id limit 25;')
rows = cur.fetchall()
conn.commit()
for row in rows:
    print(row)
# NULLIF(..., 0) avoids division by zero in case
```

```python
# Order Counts by Shipping Mode: Count order_id per ship_mode, ordered descending
cur = conn.cursor()
cur.execute('select ship_mode, count(order_id) as order_count from orders group by ship_mode order by order_count desc;')
rows = cur.fetchall()
conn.commit()
for row in rows:
    print(row)
```

```python
# Month with Highest Orders: Find the month number (MONTH(order_date)) with the greatest count of order_id
cur = conn.cursor()
cur.execute('select extract(month from order_date) as order_month, count(order_id) as order_count from orders group by order_month order by order_count desc limit 3;')
rows = cur.fetchall()
conn.commit()
for row in rows:
    print(row)
```

```python
# Top 5 Most Discounted Products: Select product_id and discount, ordered by discount descending.
cur = conn.cursor()
cur.execute('select product_id, sum(discount) as total_discount from orders group by product_id order by total_discount desc limit 5;')
rows = cur.fetchall()
conn.commit()
for row in rows:
    print(row)
```

```python
# Top 3 Cities by Sales: Sum selling_price per city, list the top three
cur = conn.cursor()
cur.execute('select city, sum(selling_price) as total_selling_price from orders group by city order by total_selling_price desc limit 3;')
rows = cur.fetchall()
conn.commit()
for row in rows:
    print(row)
```

```python
# Average Profit for 2023: Calculate the average profit for orders where YEAR(order_date)=2023
cur = conn.cursor()
cur.execute('select extract(year from order_date) as order_year, round(avg(profit),2) as average_profit from orders where extract(year from order_date) = 2023 group by order_year;')
rows = cur.fetchall()
conn.commit()
for row in rows:
    print(row)
```

```python
# Category & Sub-Category Sales: Sum selling_price for each (category, sub_category) pair.
cur = conn.cursor()
cur.execute('select category, sub_category, sum(selling_price) as total_sum_sales from orders group by category,sub_category order by total_sum_sales desc;')
rows = cur.fetchall()
conn.commit()
for row in rows:
    print(row)
```

```python
# Top 3 Profitable Products per Sub-Category: For each sub_category, list the top three product_id by total profit.
cur = conn.cursor()
cur.execute('''
    WITH ranked_products AS (
        SELECT 
            sub_category,
            product_id,
            SUM(profit) AS total_profit,
            ROW_NUMBER() OVER (PARTITION BY sub_category ORDER BY SUM(profit) DESC) AS rank
        FROM orders
        GROUP BY sub_category, product_id
    )
    SELECT sub_category, product_id, total_profit
    FROM ranked_products
    WHERE rank <= 3
    ORDER BY sub_category, total_profit DESC;
''')
rows = cur.fetchall()
conn.commit()
for row in rows:
    print(row)
```

# Students Performance in Exams Linear Regression Task

```python
# First import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# Next we load the dataset
df = pd.read_csv(r"C:\Users\user\Machine Learning\StudentsPerformance (2).csv")
df.head()
```

```python
df.isnull().sum()
```

```python
# Plot comparisons in the dataframe
plt.scatter(df['reading score'], df['math score'], color='blue')
plt.title('Math Score Vs Reading score')
plt.xlabel('Reading Score')
plt.ylabel('Math Score')
plt.grid(True)
plt.show()
```

```python
X = df[['reading score']]
y = df['math score']

model = LinearRegression()
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```

```python
df['Predicted'] = model.predict(X)

plt.scatter(df['reading score'], df['math score'], label='Actual', color='blue')
plt.plot(df['reading score'], df['Predicted'], label='Regression Line', color='red')
plt.title('Simple Linear Regression')
plt.xlabel('Reading Score')
plt.ylabel('Math Score')
plt.legend()
plt.grid(True)
plt.show()
```

EXPLANATION

```python
# Every blue dot point represents a student’s actual reading score and corresponding math score. 

# The points are tightly clustered around the red line, signifying a strong linear correlation between reading and math scores.

# It signifies that students who score higher in reading also tend to score higher in math.
```

```python
x_test = pd.DataFrame({'reading score': [80, 60, 40]})
# Predict
answer = model.predict(x_test)
print(answer)
```

```python
mae = mean_absolute_error(y, df['Predicted'])

# find mse

mse = mean_squared_error(y, df['Predicted'])

# Find rmse

rmse = np.sqrt(mse)

# Find the R2 Score (R squared)

r2 = r2_score(y, df['Predicted'])

print(f'mae: {mae}')
print(f'mse: {mse}')
print(f'rmse: {rmse}')
print(f'r2: {r2}')
```

### Detour For Q24

When your dataset has missing values (NaN), models like LinearRegression() in scikit-learn will raise an error because they expect a clean, numeric input with no gaps.

Rather than removing entire rows (which can throw away useful data), imputation lets you fill in those missing values with something reasonable — like the mean or median of the column.

Import The Imputer

```python
from sklearn.impute import SimpleImputer
```

Create the Imputer

```python
# 'mean': replaces NaNs with the mean of the column
imputer = SimpleImputer(strategy='mean')
```

Fit and Transform the Training Set

```python
# Apply the imputer to the feature matrix X_train
X_train= imputer.fit_transform(X_train)
```

Q24. Train a linear regression model.

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
df.isnull().sum()
```

** Feature Scaling: Why & How**

Q45. Why do we scale features before training a model?

A) To make the data easier to store


B) To reduce RAM usage


C) To ensure all features contribute equally


D) To remove duplicates


The correct answer is C) To ensure all features contribute equally

> -  Feature scaling is a preprocessing technique that transforms feature values to a similar scale, ensuring all features contribute equally to the model.

Q46. What is the most common scaler used in regression problems?

A) OneHotEncoder


B) MinMaxScaler


C) StandardScaler


D) LabelEncoder


The correct answer is C) StandardScaler

> - It is used to standardize the values of numeric features.

> -  It transforms data so that the mean becomes 0 and the standard deviation becomes 1 and it is ideal for algorithms like SVM, logistic regression or neural networks that assume data is normally distributed

Q48. If a feature like RAM ranges from 2 to 16 and Battery ranges from 2000 to 6000, what problem could occur if not scaled?
